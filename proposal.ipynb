```python
import requests, json

url = "https://gamma-api.polymarket.com/events?limit=5"
r = requests.get(url, timeout=20)
r.raise_for_status()
data = r.json()

print("Type:", type(data))
print("Number of items:", len(data))

# See keys of the first event (top-level fields)
print("Top-level keys:", list(data[0].keys()))

# Pretty-print the first event so you can see nested structure
print(json.dumps(data[0], indent=2)[:3000])
```

Testing Kalshi Data Loading API

```python
import requests

kalshi_url = "https://api.elections.kalshi.com/trade-api/v2/series/KXHIGHNY"
response = requests.get(kalshi_url)
series_data = response.json()

print(f"Series Title: {series_data['series']['title']}")
print(f"Frequency: {series_data['series']['frequency']}")
print(f"Category: {series_data['series']['category']}")

    
```
